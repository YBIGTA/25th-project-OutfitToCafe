```python
import json
import re

# 원본 JSON 파일 경로
input_file = '/Users/macbook/Desktop/cafe_review_summary&keyword_data/keyword_without_weight.json'

# 수정된 JSON 파일 경로
output_file = '/Users/macbook/Desktop/cafe_review_summary&keyword_data/keyword_without_weight_filtered.json'

# 제외할 카페 이름 리스트
exclude_cafes = [
    "니드인",
    "플록스",
    "언니커피",
    "ONE BEAN 카페",
    "스텀필즈커피",
    "카페수다",
    "팔롬비니",
    "카페아모렘",
    "카페조이",
    "카페리빈 성수점",
    "원스카페",
    "라니커피",
    "카페라샤워",
    "카페바로",
    "도토리커피os"
]

# 함수: 카페 이름을 정규화 (소문자로 변환, 공백과 특수 문자 제거)
def normalize_cafe_name(name):
    return re.sub(r'\s+', '', name).lower()

# 정규화된 제외할 카페 이름 리스트
exclude_cafes_normalized = [normalize_cafe_name(cafe) for cafe in exclude_cafes]

# JSON 파일 읽기
with open(input_file, 'r', encoding='utf-8') as f:
    data = json.load(f)

# JSON 파일에서 카페 이름을 모두 출력하여 확인
cafe_names = list(data.keys())
print("JSON 파일에서 발견된 카페 이름 목록:")
for cafe in cafe_names:
    print(cafe)

# 총 카페 수 계산
total_cafes = len(data)
print(f"총 {total_cafes}개의 카페가 원본 JSON 파일에 있습니다.")

# 제외할 카페 이름을 제외한 새로운 데이터를 담을 딕셔너리 생성
filtered_data = {
    cafe: keywords for cafe, keywords in data.items()
    if normalize_cafe_name(cafe) not in exclude_cafes_normalized
}

# 필터링 후 남은 카페 수 계산
filtered_cafes_count = len(filtered_data)
print(f"제외된 카페를 제외한 후 {filtered_cafes_count}개의 카페가 남았습니다.")

# 새로운 JSON 파일로 저장
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(filtered_data, f, ensure_ascii=False, indent=4)

print(f"새로운 JSON 파일이 '{output_file}'에 저장되었습니다.")
```
