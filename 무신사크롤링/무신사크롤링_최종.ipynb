```python
import sys
sys.path.append('/Users/minseo/.pyenv/versions/3.10.10/lib/python3.10/site-packages/')

import os
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from bs4 import BeautifulSoup
import time
import random
import requests
```

# 1 이미지 크롤링 - 레트로 스타일

```python
# 스타일
style = '레트로'
style_code = '18'


# Chrome driver 설정
driver = webdriver.Chrome()

url = f'https://www.musinsa.com/snap/main/recommend?brands=&category=&genders=&height-range=&seasons=&styles={style_code}&tpos=&weight-range=&sort=NEWEST'
driver.get(url)
time.sleep(2)

# 변수 설정
img_urls = []
img_alts = []

# selector (바뀜)
selector = '#commonLayoutContainer > main > main > div.sc-f98f05ee-0.ipBIJV > div.sc-aaa4dfc3-0.gIYmjL > div:nth-child(2) > div > div.snap-grid-list'

# 스크롤 중간에 정지할 시간 설정
scroll_pause_time = 0.3


collecting = True
while collecting:
    # 현재 페이지의 HTML 파싱
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    snap_grid_list = soup.select_one(selector)
    
    # 현재 로드된 모든 이미지 태그 수집
    img_tags = snap_grid_list.find_all('img', alt=True)
    new_images_count = 0
    
    for img_tag in img_tags:
        img_url = img_tag['src']
        img_alt = img_tag['alt']
        
        
        # 이미지 URL 중복 확인
        if img_url not in img_urls:
            img_urls.append(img_url)
            img_alts.append(img_alt)

            # 이미지 수가 1600을 넘으면 while구문 그만 돌도록
            if len(img_urls) >= 1600:
                print(f'##########{len(img_urls)}장의 이미지 수집 완료##########')
                collecting=False
                break
    print(f'{len(img_urls)}만큼의 이미지를 수집했습니다..')
  
    
    # 스크롤
    driver.execute_script("window.scrollBy(0, 1000);")
    time.sleep(scroll_pause_time) 

    
driver.quit()
```

```python
# 병렬저장
from concurrent.futures import ThreadPoolExecutor

# 이미지 저장 폴더 설정
if not os.path.exists(style):
    os.makedirs(style)

# 중복 방지를 위해 이미 다운로드한 이미지 URL 저장
downloaded_urls = set()

# 병렬 다운로드 함수 정의
def download_image(idx, img_url, img_alt):
    if img_url.startswith('http') and img_url not in downloaded_urls:
        img_response = requests.get(img_url)
        img_name = f'{img_alt}.jpg'  # 고유한 이름 설정
        with open(f'{style}/{img_name}', 'wb') as f:
            f.write(img_response.content)
        downloaded_urls.add(img_url)
        print(f'{idx+1}번째 이미지 다운 완료: {img_name}')
    else:
        print(f'Skipped duplicate: {img_url}')

# 병렬 다운로드 실행
with ThreadPoolExecutor(max_workers=10) as executor:
    futures = [executor.submit(download_image, idx, img_url, img_alts[idx]) for idx, img_url in enumerate(img_urls)]

# 모든 작업이 완료될 때까지 기다립니다.
for future in futures:
    future.result()

print('모든 이미지들이 다운로드되었습니다.')
```
